### Лемматизация

```python
text = df['purpose'].unique()
```

```python
text
```

```python
#from pymystem3 import Mystem
#m = Mystem() 
#lemmas = m.lemmatize(text)
#print(lemmas) 
# AttributeError: 'numpy.ndarray' object has no attribute 'splitlines'
from nltk.stem import SnowballStemmer 
russian_stemmer = SnowballStemmer('russian')  
for word in text:
    print ('Исходное слово - ' + word + ', после стемминга - ' + russian_stemmer.stem(word)) 
```

```python
house = []
auto = []
education = []
wedding = []
for query in text:
    for word in query.split():
        stemmed_word = russian_stemmer.stem(word)
        if stemmed_word == 'жил' or stemmed_word == 'недвижим':
            house.append(query) 
        if stemmed_word == 'автомобил':
            auto.append(query)
        if stemmed_word == 'образован':
            education.append(query)
        if stemmed_word == 'свадьб':
            wedding.append(query)

print(house) 
print()
print(auto)
print()
print(education)
print()
print(wedding)
```

Стало ясно, что основных целей кредита 4: автомобиль, образование, недвижимость и жилье (объеденил в одно), свадьба. Эти 4 цели понадобятся для категоризации данных.

### Категоризация данных

```python
df['purpose'].value_counts()
```

```python
def purpose_def(word):
    if 'жил' in word:
        return 'недвижимость'
    if 'недвижим' in word:
        return 'недвижимость'
    if 'автомобил' in word:
        return 'автомобиль'
    if 'образован' in word:
        return 'образование'
    if 'свадьб' in word:
        return 'свадьба'
```

```python
purpose_def('строительство жилой недвижимости')
```

```python
df['purpose_group'] = df['purpose'].apply(purpose_def)
```

```python
def total_income_def(income):
    if income < 75000:
        return 'низкий доход'
    if income < 150000:
        return 'средний доход'
    if income >= 150000:
        return 'высокий доход'
```

```python
df['total_income_group'] = df['total_income'].apply(total_income_def)
```

<div class="alert alert-block alert-info">
<b>Совет:</b> Интересно, почему ты решил разделить группы именно на такие разбиения, и получились ли группы равномерно распределенные, чтобы их можно было объективно сравнивать.
